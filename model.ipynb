# Correlation

```python
# constructing heatmap to understand the correlation
correlation = gold_data.select_dtypes(include=np.number).corr()
plt.figure(
    figsize = (8, 8)
    )
sns.heatmap(
    correlation,
    cbar = True,
    square = True,
    fmt = '.1f',
    annot = True,
    annot_kws = {
        'size': 8
        },
    cmap = 'Blues'
    )
```

```python
# correlation values of GLD
print(correlation["GLD"])
```
